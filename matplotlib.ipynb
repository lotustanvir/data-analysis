```
# Univariate - Numerical
```

```
data = {
    "salary" : [25000, 35000, 45000, 40000, 56000, 60000, 31000, 87000]
}
```

```
df = pd.DataFrame(data)
```

```
df.head()
```

```
df.shape
```

```
# line plot:
plt.plot(df["salary"], color = "red", marker = "o", lineStyles)
plt.show()
```

```
#Histogram
plt.hist(df["salary"], bins = 3, color = "red")
plt.grid()
plt.show()
```

```
df.loc[8] = [0]
```

```
#Box plot
plt.boxplot(df["salary"])
plt.show()
```

```
df.shape
```

```
#Univariate - Categorical
```

```
df["dept"] = ["HR", "HR", "IT", "IT", "IT", "IT", "HR", "IT"]
```

```
df.head()
```

```
# pie chart

count = df["dept"].value_counts()
count
```

```
plt.pie(count, labels= count.index, autopct = "%1.2f", explode = [0, 0.1])
plt.axis("equal")
plt.show()
```

```
#countplot

plt.bar(count.index, count, color = ["green", "red"])
plt.show()
```

```
#bivariate - numerical - numerical

```

```
df.head()
```

```
df["age"] = [22, 35, 24, 30, 35, 40, 36, 28]
df.head()
```

```
# Scatter plot

plt.scatter(df["salary"], df["age"], color = "red")
plt.show()

```

```
sort_age = df.sort_values("age")
sort_age
```

```
# line plot:

plt.plot(sort_age["age"],df["salary"], df["age"], color = "red", marker = "*", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```
# bar chart:

plt.bar(sort_age["age"], df["salary"], color = "red", align = "center")
plt.show()
```

```
#bivariate: Numerical - Categorical
```

```
hr_sal = df[df["dept"] == "HR"]["salary"]
hr_sal
```

```
it_sal = df[df["dept"] == "IT"]["salary"]
it_sal
```

```
# Boxplot:

plt.boxplot([hr_sal, it_sal], label= ["HR", "IT"])
plt.legend()
plt.show()
```
